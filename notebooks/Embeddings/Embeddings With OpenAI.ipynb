# Embeddings Techniques

Embeddings are a numerical representation of text that can be used to measure the relatedness between two pieces of text. Embeddings are useful for search, clustering, recommendations, anomaly detection, and classification tasks. 


https://platform.openai.com/docs/models#embeddings

```python
# load all the envinorment variables
from dotenv import load_dotenv
load_dotenv()
```

```python
import os
os.environ['OPENAI_API_KEY'] = os.getenv("OPENAI_API_KEY")
```

```python
from langchain_openai import OpenAIEmbeddings
openai_embedding = OpenAIEmbeddings(model='text-embedding-3-large')
openai_embedding
```

```python
text = "Exploring Embedding Techniques Using OpenAI with LangChain"
# Generate embeddings
vector = openai_embedding.embed_query(text)
```
